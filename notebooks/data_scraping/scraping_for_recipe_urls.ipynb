```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_recipes(base_url, start_page, end_page):
    # Header untuk menghindari blokir
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }

    all_links = []

    for page in range(start_page, end_page + 1):
        url = f"{base_url}/page/{page}/"
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Mencari semua elemen yang mengandung tautan resep
        recipes = soup.find_all('article', class_='post-summary')
        
        # Menyimpan tautan resep
        links = []
        for recipe in recipes:
            link = recipe.find('a')['href']
            links.append(link)

        # Mengabaikan 6 tautan terakhir
        if len(links) > 6:
            links = links[:-6]

        all_links.extend(links)
    
    return all_links
```

```python
# URL dasar
base_url = "https://addapinch.com/category/recipes-by-course/desserts/"
```

```python
# Memanggil fungsi scrape_recipes
recipe_links = scrape_recipes(base_url, 2, 17)
```
